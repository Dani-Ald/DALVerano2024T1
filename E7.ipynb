```python
#IMPORTACIONES
import Caracteres.abs as frecuencia_absoluta
import Caracteres.calcular as calcular_rango
import Caracteres.calcular as calcular_numero_clases
import Caracteres.calcular as calcular_ancho_clase
import Caracteres.calcular as calcular_limites_clase
import Caracteres.calcular as calcular_marcas_clase
import Caracteres.graficos as crear_histograma
import Caracteres.graficos as crear_diagrama_barras
import Caracteres.graficos as crear_poligono_frecuencias
import Caracteres.graficos as crear_grafico_pastel
import Caracteres.graficos as crear_ojiva
import Caracteres.imprimirtabla as imprimir_tabla
import Caracteres.ordendatos as ordenar_datos
import Caracteres.relativa as frecuencia_relativa
import Caracteres.acumulada as frecuencia_acumulada

# USO DE LA FUNCION
datos = [
    "Pumas", "América", "Chivas", "Pumas", "Chivas",
    "Chivas", "América", "Toluca", "Toluca", "Pumas",
    "Santos", "Necaxa", "América", "América", "América",
    "Pumas", "Santos", "Pumas", "Pumas", "Pumas",
    "América", "Chivas", "América", "Pumas", "Necaxa",
    "Toluca", "Chivas", "América", "Toluca", "Santos"
]

# Paso 1: Ordenar los datos por longitud de cadena
datos_ordenados = ordenar_datos.ordenar_datos(datos)

# Paso 2: Calcular el rango (basado en la longitud de las cadenas)
rango, valor_min, valor_max = calcular_rango.calcular_rango(datos_ordenados)

# Paso 3: Determinar el número de clases
n = len(datos_ordenados)
num_clases = calcular_numero_clases.calcular_numero_clases(n)

# Paso 4: Calcular el ancho de clase
ancho_clase = calcular_ancho_clase.calcular_ancho_clase(rango, num_clases)

# Paso 5: Calcular límites de clase (convertir longitud de cadenas en rangos numéricos)
limites_inferiores, limites_superiores = calcular_limites_clase.calcular_limites_clase(valor_min, ancho_clase, num_clases)
limites_superiores[-1] = len(valor_max)  # Asegurar que el último límite superior sea la longitud máxima

# Paso 6: Calcular marcas de clase (promedio de los límites)
marcas_clase = calcular_marcas_clase.calcular_marcas_clase(limites_inferiores, limites_superiores)

# Paso 7: Calcular frecuencias absolutas
frec_absoluta = frecuencia_absoluta.frecuencia_absoluta(datos_ordenados, limites_inferiores, limites_superiores)

# Paso 8: Calcular frecuencias relativas
frec_relativa = frecuencia_relativa.frecuencia_relativa(frec_absoluta, n)

# Paso 9: Calcular frecuencias acumuladas
frec_acumulada = frecuencia_acumulada.frecuencia_acumulada(frec_absoluta)

# Imprimir la tabla
imprimir_tabla.imprimir_tabla(datos_ordenados, limites_inferiores, limites_superiores, marcas_clase, frec_absoluta, frec_relativa, frec_acumulada)

# Crear los gráficos
crear_diagrama_barras.crear_diagrama_barras(marcas_clase, frec_absoluta)
crear_grafico_pastel.crear_grafico_pastel(frec_absoluta, marcas_clase)
```
